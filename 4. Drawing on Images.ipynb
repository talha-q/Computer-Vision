#Drawing on Images

First let's import the libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

Let's move to further process. 
Here we are going to use the image function we made in 1st and 2nd lectures

```python
def imshow (title = 'image', image=None, size=10):
  w,h=image.shape[0],image.shape[1]
  aspect_ratio=w/h
  plt.figure(figsize=(size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

##Making Black Images

###Creating Black Image in RGB

```python
image=np.zeros((512,512,3),np.uint8) #it creates a black image in RGB format 
imshow('Black in RGB',image)
```

###Creating a Black Image in Grayscale

```python
image_gray= np.zeros((512,512),np.uint8)
imshow('Black in Grayscale',image_gray)
```

##Creating Shapes
###Creating Line 
cv2.line(image, starting coodinate, ending coordinate, color, thickness) 

```python
cv2.line(image,(50,50),(450,450),(249,249,151),5) #line in Electric Blue (I just foud out that here it also goes BGR instead of RGB)
imshow('Line on RGB Image',image)
```

###Drawing Rectangles
cv2.rectangle(image, (starting vertex),(ending vertex), (Color),Thickness)

```python
#Lets draw a rectangle
cv2.rectangle(image, (150,150),(300,315),(0,219,151),5)
imshow('Rectangle on Image',image)
```

###Drawing Circles
cv2.circle(image,(center),radius, (color), fill)
Here in fill -1 indicates fill and 1 indicates outline


```python
cv2.circle(image,(250,250), 80,(175,255,255),-1)
imshow("Circle on Image",image)
```

###Polygon on Image
cv2.polylines(image, [points], Closed?(True or False),Color, thickness)


points here are reshaped array of points

```python
points=np.array([[10,50],[400,50],[90,200],[50,500]])
points=points.reshape((-1,1,2))#dont think about this step much, it just changes the format of points

cv2.polylines(image,[points],True,(190,220,110),7)
imshow("Polygon on Image", image)
```
