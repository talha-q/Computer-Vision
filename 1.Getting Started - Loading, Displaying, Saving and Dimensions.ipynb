##Displaying Images
We use Matplot library for that (But there is a little catch.... Images appear as BGR in it, we need to convert them into RGB for better display)

Lets show image with matplot library

```python
from matplotlib import pyplot as plt
```

show images with matplot

```python
plt.imshow(image)
plt.show()
```

Here the result is in BGR format as discusses earlier (this is the natural format in which matplot library sees the images)
Let's now convert it in RGB format

Let's create a function to make our lives easier

###BGR2RGB

```python
def imshow(title="",image=None):
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

the function above, named as imshow displays title of the image above,
converts image from BGR format to RGB format and displays it.

Let's see it in action

```python
imshow("Displaying our First Image",image)
```
