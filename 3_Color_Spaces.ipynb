##Working with Bands

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

#function preparation
def imshow(title="title to be", image=None, size=15):
  w,h = image.shape[0], image.shape[1]
  aspect_ratio=w/h
  plt.figure(figsize=(size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

Testing Image


```python
image = cv2.imread('./yami1.jpg')
plt.imshow(image)
```

testing image in imshow function

```python
imshow('Test RGB Image',image)
#here our function has converted the BGR image to RGB image 
#its size is bigger as I have played with size in this code (I have used size 15 in this case)
```

###Splitting Image in Individual Channels
Here we are going to use cv2.split(image) command to split the image in its individual channels

```python
B,G,R= cv2.split(image)
print(B.shape)
print(G.shape)
print(R.shape)
```

After splitting, each image band will appear as a gray scale image 


```python
plt.imshow(B)
#this shows only Blue band color values
```

```python
plt.imshow(G)
#this shows only Green band color values
```

```python
plt.imshow(R)
#this shows only Red band color values
```

It looks like a gray scale image as it contains dimension of single band

###Visualizing Individual Bands

To visualize bands we create a blank matrix of zeros

```python
zeros = np.zeros(image.shape[:2],dtype="uint8")
plt.imshow(zeros) #here is the sample zero matrix
```

```python
imshow("Red",cv2.merge([zeros,zeros,R]))
imshow("Green",cv2.merge([zeros,G,zeros]))
imshow("Blue",cv2.merge([B,zeros,zeros]))
```

###Merging Bands/Channels Back
1. Just like we did cv2.split(image) for splitting the image in its bands
2. We can now do cv2.merge to merge the bands in the image

```python
merged=cv2.merge([B,G,R])
imshow("Merged",merged)
```

The benefit of this ability is that we can enhance a particular band if we want to. For example

```python
enhancedB= cv2.merge([B+100,G,R])
enhancedG= cv2.merge([B,G+100,R])
enhancedR= cv2.merge([B,G,R+100])
imshow("enhanced B",enhancedB)
imshow("enhanced G",enhancedG)
imshow("enhanced R", enhancedR)
imshow("Normal",merged)
```
