##Preparation

3. Upload Images (I have uploaded them manually, however they can be downloaded form google drive and uploaded and unzipped there as well)


```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

#function preparation
def imshow(title="title to be", image=None, size=15):
  w,h = image.shape[0], image.shape[1]
  aspect_ratio=w/h
  plt.figure(figsize=(size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

Testing Image


```python
image = cv2.imread('./yami1.jpg')
plt.imshow(image)
```

testing image in imshow function

```python
imshow('Test RGB Image',image)
#here our function has converted the BGR image to RGB image 
#its size is bigger as I have played with size in this code (I have used size 15 in this case)
```

##Splitting Image in Individual Channels
Here we are going to use cv2.split(image) command to split the image in its individual channels

```python
B,G,R= cv2.split(image)
print(B.shape)
print(G.shape)
print(R.shape)
```
