#Grayscaling Images

```python
import cv2
from matplotlib import pyplot as plt
```

Loading Image (for better understanding refer to lec 1)

```python
image = cv2.imread('./temp.jpg')
```

Displaying image (this will appear in BGR rather than RGB as this is the default orientation as shown in lec 1)

```python
plt.imshow(image) #demo check of image
```

imgshow function

```python
def imshow(title="Any_Title", image=None, size=10):
  w,h=image.shape[0],image.shape[1]
  aspect_ratio=w/h
  plt.figure(figsize=(size*aspect_ratio,size))
  plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
  plt.title(title)
  plt.show()
```

##Loading Image

```python
image=cv2.imread('temp.jpg')
```

```python
imshow("Lake SaifulMalook",image)
```

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
```

```python
imshow('Saif-ul-Malook Converted to Gray Scale', gray_image)
```
